```python
# Annuity Plan Calculation

# Loop for Payment Amount (PMT)
while True:
    PMT = input("Enter the Payment per period (PMT): ")
    try:
        PMT = float(PMT)
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value.")

# Loop for Rate (R)
while True:
    R = input("Enter the Rate of interest (R) in percentage: ")
    try:
        R = float(R)
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value.")

# Loop for Number of Periods (n)
while True:
    n = input("Enter the number of payment periods (n): ")
    try:
        n = int(n)
        if n <= 0:
            print("Please enter a positive whole number.")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter a whole number.")

# Calculate Annuity Amount
A = PMT * (((1 + (R / 100))**n - 1) / (R / 100))
print("Total Amount from Annuity Plan is:", A)
```
