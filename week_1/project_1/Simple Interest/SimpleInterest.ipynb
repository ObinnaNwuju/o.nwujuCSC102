```python
# Loop for Principal (P)
while True:
    P = input("Enter the Principal amount (P): ")
    try:
        P = float(P)  # attempt conversion to number
        break  # exit loop if successful
    except ValueError:
        print("Invalid input. Please enter a numerical value.")
        continue

# Loop for Rate (R)
while True:
    R = input("Enter the Rate of interest (R) in percentage: ")
    try:
        R = float(R)
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value.")
        continue

# Loop for Time (T)
while True:
    T = input("Enter the Time period in years (T): ")
    try:
        T = float(T)
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value.")
        continue

# Calculate Simple Interest
I = (P * R * T) / 100
print("Simple Interest is:", I)
```
