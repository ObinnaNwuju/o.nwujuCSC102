```python
# Compound Interest Calculation

# Loop for Principal (P)
while True:
    P = input("Enter the Principal amount (P): ")
    try:
        P = float(P)  # convert to float
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value.")

# Loop for Rate (R)
while True:
    R = input("Enter the Rate of interest (R) in percentage: ")
    try:
        R = float(R)
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value.")

# Loop for Time (T)
while True:
    T = input("Enter the Time period in years (T): ")
    try:
        T = float(T)
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value.")

# Loop for Compounding Frequency (n)
while True:
    n = input("Enter the number of times interest is compounded per year (n): ")
    try:
        n = int(n)
        if n <= 0:
            print("Please enter a positive whole number.")
            continue
        break
    except ValueError:
        print("Invalid input. Please enter a whole number.")

# Calculate Compound Interest
A = P * (1 + (R / (100 * n))) ** (n * T)
print("Final Amount with Compound Interest is:", A)
```
