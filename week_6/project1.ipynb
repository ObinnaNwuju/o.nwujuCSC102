```python
import tkinter as tk
from tkinter import messagebox
import random

class Employee:
    def __init__(self):
        self.employees = [
           "MARY EVANS", "EYO ISHAN", "DUROJAIYE DARE", "ADAMS ALI",
"ANDREW UGWU", "STELLA MANKINDE", "JANE AKIBO", "AGO JAMES",
"MICHELL TAIWO", "ABRAHAM JONES", "NICOLE ANIDE", "KOSI KORSO",
"ADELE MARTINS", "EMMANUEL OJO", "AJAYI FATIMA"
        ]

        self.tasks = [
            "Loading", "Transporting", "Reviewing Orders",
            "Customer Service", "Delivering Items"
        ]

   
    def check_employee(self, name):
        return name in self.employees

 
    def take_attendance(self, name):
        return f"Attendance has been taken for {name}"

  
    def assign_task(self):
        randomnumber = random.randint(0, len(self.tasks) - 1)
        return self.tasks[randomnumber]

  
    def refuse_access(self):
        return "Access Denied. You are not an employee."


system = Employee()

def submit():
    name = entry.get()

    if system.check_employee(name):
        attendance = system.take_attendance(name)
        task = system.assign_task()

        messagebox.showinfo(
            "Access Granted",
            attendance + "\nTask: " + task
        )
    else:
        messagebox.showerror(
            "Denied",
            system.refuse_access()
        )


root = tk.Tk()
root.title("Employee System")
root.geometry("500x300")

tk.Label(root, text="Enter Employee Name (All caps)").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Submit", command=submit).pack()

root.mainloop()
```
