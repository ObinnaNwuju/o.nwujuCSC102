```python
import tkinter as tk # mport librareis
from tkinter import messagebox 

class DeliveryService: # main class for the business logic
    def __init__(self, root): # constructor to set up the window
        self.root = root 
        self.root.title("Jane's Delivery - Price Calculator") 
        self.root.geometry("350x300") 
        
        # location entry
        self.loc_label = tk.Label(root, text="Enter Location (PAU or Epe):")
        self.loc_label.pack(pady=5)
        self.location_entry = tk.Entry(root)
        self.location_entry.pack(pady=5)

        # weight entry
        self.weight_label = tk.Label(root, text="Enter Weight (only numbers eg 10,15) (kg):")
        self.weight_label.pack(pady=5)
        self.weight_entry = tk.Entry(root)
        self.weight_entry.pack(pady=5)

        # calculate Button
        self.calc_button = tk.Button(root, text="Calculate Price", command=self.calculate_price, bg="green", fg="white")
        self.calc_button.pack(pady=20)

    def calculate_price(self): 
        #get input
        location = self.location_entry.get().strip().upper()
        weight = float(self.weight_entry.get())
      
        price = 0 # initialize price \

        if location == "PAU":
            if weight >= 10: # 10kg and above
                price = 2000
            else: # less than 10kg
                price = 1500
                
        elif location == "EPE":
            if weight >= 10: # 10kg and above
                price = 5000
            else: # less than 10kg
                price = 4000
                
        else: # if the location isnt PAU or Epe
            messagebox.showwarning("Notice", "We only deliver to PAU and Epe.")
            return

   
        messagebox.showinfo("info", f"The delivery charge to {location.title()} for {weight}kg is: N{price}")


root = tk.Tk() 
app = DeliveryService(root) 
root.mainloop() 
```
