```python
import cmath  # for complex number support
import numpy as np  # for cubic and quartic roots

def quadratic_roots(a, b, c):
    d = cmath.sqrt(b**2 - 4*a*c)
    root1 = (-b + d)/(2*a)
    root2 = (-b - d)/(2*a)
    return root1, root2

def cubic_roots(a, b, c, d):
    # coefficients in descending order
    coeffs = [a, b, c, d]
    return np.roots(coeffs)  # returns all roots (real and complex)

def quartic_roots(a, b, c, d, e):
    coeffs = [a, b, c, d, e]
    return np.roots(coeffs)

# Main program
print("Select operation:")
print("1. Quadratic Equation")
print("2. Cubic Equation")
print("3. Quartic Equation")

choice = input("Enter choice (1/2/3): ")

if choice == "1":
    a = float(input("Enter coefficient A: "))
    b = float(input("Enter coefficient B: "))
    c = float(input("Enter coefficient C: "))
    r1, r2 = quadratic_roots(a, b, c)
    print("Roots of the quadratic equation are:", r1, "and", r2)

elif choice == "2":
    a = float(input("Enter coefficient A: "))
    b = float(input("Enter coefficient B: "))
    c = float(input("Enter coefficient C: "))
    d = float(input("Enter coefficient D: "))
    roots = cubic_roots(a, b, c, d)
    print("Roots of the cubic equation are:", roots)

elif choice == "3":
    a = float(input("Enter coefficient A: "))
    b = float(input("Enter coefficient B: "))
    c = float(input("Enter coefficient C: "))
    d = float(input("Enter coefficient D: "))
    e = float(input("Enter coefficient E: "))
    roots = quartic_roots(a, b, c, d, e)
    print("Roots of the quartic equation are:", roots)

else:
    print("Invalid choice")
```
