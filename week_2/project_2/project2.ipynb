```python
import numpy as np

def find_roots(coefficients):
    """
    Return all roots (including complex) of a polynomial given its coefficients.
    Coefficients are ordered from highest degree to constant term.
    """
    return np.roots(coefficients)

def main():
    print("=== Root Finder for Polynomial Equations ===")
    print("1. Quadratic equation  (ax² + bx + c = 0)")
    print("2. Cubic equation      (ax³ + bx² + cx + d = 0)")
    print("3. Quartic equation    (ax⁴ + bx³ + cx² + dx + e = 0)")
    
    choice = input("\nSelect operation (1/2/3): ").strip()
    
    if choice == '1':
        print("\n--- Quadratic Equation ---")
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        coeffs = [a, b, c]
    elif choice == '2':
        print("\n--- Cubic Equation ---")
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        d = float(input("Enter coefficient d: "))
        coeffs = [a, b, c, d]
    elif choice == '3':
        print("\n--- Quartic Equation ---")
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        d = float(input("Enter coefficient d: "))
        e = float(input("Enter coefficient e: "))
        coeffs = [a, b, c, d, e]
    else:
        print("Invalid choice. Please run the program again.")
        return
    
    # Check that the leading coefficient is not zero
    if coeffs[0] == 0:
        print("Error: Leading coefficient cannot be zero for the selected degree.")
        return
    
    # Compute roots
    roots = find_roots(coeffs)
    
    # Display results
    print("\n" + "="*40)
    print("Roots found:")
    for i, r in enumerate(roots, 1):
        # Format complex numbers nicely
        if np.iscomplexobj(r) and np.imag(r) == 0:
            r = np.real(r)  # show as real if imaginary part is zero
        print(f"Root {i}: {r}")
    print("="*40)

if __name__ == "__main__":
    main()
```
