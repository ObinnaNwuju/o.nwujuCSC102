```python
def age_experience():
    age = int(input("Please tell us your age: "))
    experience = int(input("How many years have you worked for?: "))

    while True:
        if age >= 55 and experience > 25:
            print("Your Average Tax Revenue is N5,600,600")
            break
        elif age >= 45 and experience >= 20:
            print("Your Average Tax Revenue is N4,480,000")
            break
        elif age >= 35 and experience >= 10:
            print("Your Average Tax Revenue is N1,500,000")
            break
        elif age < 35 and experience < 10:
            print("Your Average Tax Revenue is N550,000")
            break
            

age_experience()
```
