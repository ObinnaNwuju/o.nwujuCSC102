def checkout():
    print("\n=== Store Checkout ===")
    total = 0.0

    while True:
        try:
            price = float(input("Enter item price (or 0 to finish): "))
        except ValueError:
            print("Please enter a valid number.")
            continue
        if price == 0:
            break
        total += price

    print(f"Subtotal: ${total:.2f}")

    while True:
        try:
            payment = float(input("Enter payment amount: "))
        except ValueError:
            print("Please enter a valid number.")
            continue
        if payment >= total:
            break
        print(f"Insufficient payment. You need at least ${total:.2f}.")

    change = payment - total
    print(f"Change: ${change:.2f}")
    print("Thank you for shopping!\n")

if __name__ == "__main__":
    checkout()

```python
def swap_ages():
    """Swap the ages of two people."""
    print("\n=== Swap Ages ===")
    name1 = input("Enter first person's name: ")
    age1 = int(input(f"Enter {name1}'s age: "))
    name2 = input("Enter second person's name: ")
    age2 = int(input(f"Enter {name2}'s age: "))

    # Swap ages
    age1, age2 = age2, age1

    print(f"After swapping: {name1} is now {age1} years old, {name2} is now {age2} years old\n")

if __name__ == "__main__":
    swap_ages()
```
