# Store Checkout Program

total_cost = 0  # initialize total cost

# Ask how many items the user wants to purchase
while True:
    total_items = input("Enter the total number of items you wish to purchase: ")
    if total_items.isdigit():
        total_items = int(total_items)
        break
    else:
        print("Invalid input. Please enter a number.")

# Loop for each item
for i in range(total_items):
    while True:
        price = input(f"Enter price of item {i+1}: ")
        quantity = input(f"Enter quantity of item {i+1}: ")

        if not price.isdigit() or not quantity.isdigit():
            print("Invalid input. Please enter numbers only.")
            continue
        else:
            price = float(price)
            quantity = int(quantity)
            break

    total_cost += price * quantity

print(f"\nTotal amount due: {total_cost}")

# Payment input
while True:
    payment = input("Enter payment amount: ")
    if payment.replace('.', '', 1).isdigit():  # allow decimals
        payment = float(payment)
        if payment < total_cost:
            print(f"Insufficient payment. You need at least {total_cost}.")
            continue
        else:
            break
    else:
        print("Invalid input. Enter a number.")

change = payment - total_cost
print(f"Payment received: {payment}")
print(f"Change to return: {change}")
print("Receipt printed. Thank you for shopping!")
